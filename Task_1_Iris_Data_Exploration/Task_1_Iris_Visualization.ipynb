# Task 1: Exploring and Visualizing the Iris Dataset

```python
# import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the iris datset directly from seaborn
iris=sns.load_dataset('iris')
```

```python
# use head() -> to print first 5 rows by default
print(iris.head())
```

```python
# print info about datatypes and missing value
iris.info()
```

```python
# describe -> describe numerical information about each column
iris.describe()
```

```python
# sum to missing values in each column
iris.isnull().sum()
```

```python
# Create scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=iris,x="sepal_length",y="petal_length",hue="species",style="species")
plt.title("Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.grid(True)
plt.show()
```

```python
# Plot histograms for each numeric column
iris.hist(
    figsize=(10, 8),          # Size of entire figure
    edgecolor='black',        # Border for bars
    color='skyblue'           # Bar color
)

plt.suptitle(" Histogram of Iris Features", fontsize=16)
plt.tight_layout()
plt.show()
```
